# 1: Elderly Pupolation rank

Get data for the population by 5-year group in SUA

```python
df = pd.read_csv('Cleaned_Age_in_Five_Year_Groups_in_SUA.csv', header = 0, names = ['cities', '0-4_years','5-9_years','10-14_years', '15-19_years', '20-24_years', '25-29_years', '30-34_years', '35-39_years', '40-44_years', '45-49_years', '50-54_years', '55-59_years', '60-64_years', '65-69_years', '70-74_years', '75-79_years', '80-84_years', '85-89_years', '90-94_years', '95-99_years', '100_years_and_over', 'total']);
# df
```

Filter the SUA in NSW

```python
def get_state_from_city(city):
    location = geolocator.geocode(city, exactly_one=True)
    if location:
        if (location.raw['address_components'][1]['types'][0] == 'administrative_area_level_1'):
            return location.raw['address_components'][1]['short_name']
        elif (location.raw['address_components'][1]['types'][0] == 'administrative_area_level_2'):
            return location.raw['address_components'][2]['short_name']
    else:
        return None

# get_state_from_city('Sydney')
```

```python
df['state'] = df['cities'].apply(lambda x: get_state_from_city(x.split(' - ')[0]))
df = df.loc[df['state'] == 'NSW']
df
```

```python
df.loc[:, '65_years_and_over_total'] = (
    df['65-69_years'] + df['70-74_years'] + df['75-79_years'] +
    df['80-84_years'] + df['85-89_years'] + df['90-94_years'] +
    df['95-99_years'] + df['100_years_and_over']
)

df = df.drop(columns=[
    '0-4_years', '5-9_years', '10-14_years', '15-19_years', '20-24_years',
    '25-29_years', '30-34_years', '35-39_years', '40-44_years', '45-49_years',
    '50-54_years', '55-59_years', '60-64_years', '65-69_years', '70-74_years',
    '75-79_years', '80-84_years', '85-89_years', '90-94_years', '95-99_years',
    '100_years_and_over'
])

df = df.loc[:, ['cities', '65_years_and_over_total', 'total', 'state']]
```

```python
df = df.drop(columns=['state'])
```

## 1-1: Rank for total population

```python
df = df.sort_values('total', ascending = False)

# reset index after sorting
df = df.reset_index(drop = True)

# create a new column 'Total_Rank'
df['total_rank'] = df.index + 1

# calculate the log value
df['total_population_log'] = np.log10(df['total']) 
df['total_rank_log'] = np.log10(df['total_rank'])
df
```

```python
# least squares polynomial fit
model = np.polyfit(df['total_population_log'], df['total_rank_log'], 1)
print(model)
```

```python
sns.regplot(data = df, x = df['total_population_log'], y = df['total_rank_log'],
            label='y = ' + str(round(model[0],4)) + 'x' + str(round(model[1],4)))
plt.legend()
plt.show()
```

```python
plt.bar(df['cities'], df['total'])

plt.xticks(rotation=90)
plt.show()
```

## 1-2: Rank for elderly population

Get the total for age over 65

```python
df = df.sort_values('65_years_and_over_total', ascending = False)

# reset index after sorting
df = df.reset_index(drop = True)

# create a new column 'Elderly_Rank'
df['elderly_rank'] = df.index + 1

# calculate the log value
df['elderly_population_log'] = np.log10(df['65_years_and_over_total']) 
df['elderly_rank_log'] = np.log10(df['elderly_rank'])
df
```

```python
# least squares polynomial fit
model = np.polyfit(df['elderly_rank_log'], df['elderly_population_log'], 1)
print(model)
```

```python
sns.regplot(data = df, x = df['elderly_population_log'], y = df['elderly_rank_log'],
            label='y = ' + str(round(model[0],4)) + 'x' + str(round(model[1],4)))
plt.legend()
plt.show()
```

```python
plt.bar(df['cities'], df['65_years_and_over_total'])

plt.xticks(rotation=90)
plt.show()
```

Elderly Population vs Total Population

```python
df['elderly_ratio'] = df['total'] / df['65_years_and_over_total']
df = df.sort_values('elderly_ratio', ascending = False)

# # reset index after sorting
df = df.reset_index(drop = True)

# # create a new column 'elderly_ratio_rank'
df['elderly_ratio_rank'] = df.index + 1

# # calculate the log value
df['elderly_ratio_population_log'] = np.log10(df['elderly_ratio']) 
df['elderly_ratio_rank_log'] = np.log10(df['elderly_ratio_rank'])
df
```

```python
# least squares polynomial fit
model = np.polyfit(df['elderly_ratio_rank_log'], df['elderly_ratio_population_log'], 1)
print(model)
```

```python
sns.regplot(data = df, x = df['elderly_ratio_population_log'], y = df['elderly_ratio_rank_log'],
            label='y = ' + str(round(model[0],4)) + 'x' + str(round(model[1],4)))
plt.legend()
plt.show()
```
