# 1. Introduction

In many major metropolitan areas, the aging and increasingly elderly population is becoming a prominent issue. However, these cities' high-paced and development-driven objectives often clash with the slower, retirement-oriented lifestyle favoured by older residents. As a result, a critical question arises: What characteristics define a city that is truly conducive to the needs and preferences of its elderly citizens?

Our investigation focuses on three essential factors: the proportion of the elderly population, the availability of elder care facilities, and the economic impact of these variables on elderly living. This report aims to not only describe the current landscape but also stimulate discussion and action toward enhancing cities for the benefit of our senior citizens.

# 2. Construction of the Indicator System

In order to assess the issue of the friendliness of a city for older people, we have developed a comprehensive indicator composed of several sub-indicators called the Elderly Friendly Index (EFI) to assess how friendly the city is for the elderly people. This is a comprehensive index, and our sub-indicators are broadly grouped into 3 categories: Demographic Environment, Social Facilities, and Socio-Economic, and each component is assigned a different weighting within the Master Index, based on a study of the WHO recommendations and government reports. The calculation of the master indicator will be based on the weighted composite index method, where the standardisation of each sub-indicator will be followed by the creation of a comprehensive evaluation of the city's elderly-friendly index based on the weights established.

## 2-1: Demographic Environment
There is only one sub-indicator under this classification, which is the proportion of the urban elderly population in the total population. Different percentages of the population in different age groups will constitute different urban human environments. Cities with a higher proportion of the elderly population are likely to be more conducive to creating a social atmosphere that cares for and serves the elderly, generating more social activities and cultural and recreational programs for the elderly. It also has an impact on the future infrastructure needs, economic development, and employment environment of the city.

## 2-2: Social Facilities
Improving infrastructure support is an important physical component of an elderly-friendly city. In this class of indicators, we mainly examine whether the supply of urban medical and elderly services can meet the needs of the elderly population.

### 2-2-1：Aged care service（per capita)
The number of aged care facilities per capita can reflect a city's ability to provide aged care services. Cities with an insufficient number of aged care facilities may not be able to meet the aged care needs of their elderly population. Also, aged care facilities are an important space for the social life of older people, where they have more opportunities to engage in social activities and build social relationships.

### 2-2-2 Medical Facilities(per capita)
The per capita number of hospitals within a city can serve as a representative indicator of the city's healthcare accessibility. A higher ratio of hospitals per person signifies a stronger capacity of the city to offer timely and efficient medical assistance to its elderly population, particularly in matters related to health and personal care. This underscores the city's commitment to maintaining a comprehensive and effective healthcare infrastructure to support the wellbeing of its ageing residents.

## 2-3 Social Economic
Socio-economic indicators provide an insight into the stability and security of seniors' standard of living. Parameters such as a city's average pension, government subsidies, and senior services reflect the city's capacity to furnish elderly care services and its extent of economic support for this demographic. Meanwhile, the cost of living index can serve as a barometer for the economic pressures associated with retirement life in an urban setting. Together, these factors constitute a comprehensive overview of the socio-economic landscape for the elderly. 
          
### 2-3-1 Income
Funds serve as a critical safety net, ensuring that the elderly can meet their basic needs. These resources can be derived from various channels including family income, pensions, and other retirement benefits. In this particular analysis, our primary focus rests on two components: personal income from family members and pensions. These sources of income constitute the foundation that supports daily living expenses. Furthermore, they secure a reasonable standard of living for the elderly post-retirement, thus emphasising their role as a vital component in the social security apparatus for the ageing population.

### 2-3-2 Government Support
Appropriate government financial support serves as a robust security measure for the elderly population. Through providing financial aid and welfare subsidies, the government can significantly alleviate the economic strain experienced by seniors. This targeted allocation of governmental resources not only fosters the development of the senior services industry, improving the overall quality of services, but also underpins the future construction and enhancement of elderly service facilities. This strategic investment ensures the fulfilment of basic living necessities for seniors, leading to an uplift in their quality of life and fostering a more inclusive urban environment.

### 2-3-3 Living Cost
The cost of living and the quality of pension services have a direct impact on the financial needs of the elderly population. In cities where these costs are high, the economic burden on seniors increases significantly. Consequently, it becomes incumbent upon governments and social institutions to provide additional welfare measures and subsidies. These aid initiatives help alleviate the financial stress experienced by the elderly and ensure their access to basic life necessities and retirement services. Through such mechanisms, a robust safety net for the elderly population can be maintained, promoting a sustainable and inclusive urban environment.


# 3. Indicators Table 

Our analytical framework apportions weight to these categories as follows: 10% to population demographics, 35% to social infrastructure, and 45% to socio-economic factors. Utilising this structured approach, we have generated a detailed ranking system that portrays the degree of elderly friendliness for each examined area. 


**Normalization formula：**

Score = 100 * (Value - MinValue) / (MaxValue - MinValue)

```python
def addIndicatorScore(column_name, indicatorTable, indicatorsTable, score_column = 'score'):
    if column_name in indicatorsTable.columns:
        # If it exists, drop it
        indicatorsTable.drop(columns=[column_name], inplace=True)
    return pd.merge(indicatorsTable, indicatorTable[['LGA_CODE22', score_column]], on='LGA_CODE22', how='left').rename(columns={score_column: column_name})
```

```python
import re
def currency_to_number(currency_str):
    try:
        num_str = re.sub(r'[^\d.]', '', currency_str)  # Remove non-numeric characters
        num = float(num_str)
        return num
    except ValueError:
        return None
```

```python
gdf_age_care_GS['funding'] = gdf_age_care_GS['funding'].apply(currency_to_number)

# Get the count of Age care services in each city
sum_counts = gdf_age_care_GS.groupby('LGA_CODE22').size().reset_index(name='sum_counts')

# Sum up the currency numbers for each city
sum_currency = gdf_age_care_GS.groupby('LGA_CODE22')['funding'].sum().reset_index(name='sum_funding')

# Create a new DataFrame from gdf_lga_NSW
df_age_care_in_city = pd.merge(gdf_lga_population[['LGA_CODE22', 'LGA_NAME22', '65 years and over']], sum_counts, on='LGA_CODE22', how='left')
df_age_care_in_city = pd.merge(df_age_care_in_city, sum_currency, on='LGA_CODE22', how='left')

# Fill NaN values with 0 (assuming that NaN means there are no Age care services in that local government area)
df_age_care_in_city['sum_counts'].fillna(0, inplace=True)
df_age_care_in_city['sum_funding'].fillna(0, inplace=True)

# Calculate the per capita values
df_age_care_in_city['per capital'] = df_age_care_in_city['sum_counts'] / df_age_care_in_city['65 years and over']

# Sort the dataframe by per capita count in descending order
df_age_care_in_city = df_age_care_in_city.sort_values('per capital', ascending=False).reset_index(drop=True)
print(len(df_age_care_in_city))
df_age_care_in_city
```

```python
def iterate_coordinates(city, place_type):
    # Geocode the city name to get its coordinates
    geocode_result = gmaps.geocode(city, components={'country': 'AU'})
    
    if geocode_result:
        # Extract the latitude and longitude from the geocode result
        lat = geocode_result[0]['geometry']['location']['lat']
        lng = geocode_result[0]['geometry']['location']['lng']
        
        for i in range(-2, 2):
            for j in range(-2, 2):
                new_lat = lat + (i * 0.1)
                new_lng = lng + (j * 0.1)
                
                # Call the function to search for locations
                search_location(place_type, new_lat, new_lng)
    else:
        print("Unable to geocode the provided city.")
```

```python
df_other_healthcare = pd.DataFrame(columns=['LGA_NAME22', 'lat', 'lon'])
```

```python
overpass_url = "http://overpass-api.de/api/interpreter"
def search_dentists_in_city(city, type):
    overpass_query = f"""
    [out:json];
    area["name"="{city}"][boundary=administrative];
    (node["amenity"="{type}"](area);
     way["amenity"="{type}"](area);
     rel["amenity"="{type}"](area);
    );
    out center;
    """
    response = requests.get(overpass_url, params={'data': overpass_query})
    data = response.json()
    elements = data.get('elements', [])
    for element in elements:
        lat = element['lat']
        lon = element['lon']
        df_other_healthcare.loc[len(df_other_healthcare)] = [city, lat, lon]
```

```python
for index, row in gdf_lga_NSW.iterrows():
    city_name = row['LGA_NAME22']
    search_dentists_in_city(city_name, 'dentist')
    search_dentists_in_city(city_name, 'nursing_home')
    search_dentists_in_city(city_name, 'clinic')
    search_dentists_in_city(city_name, 'pharmacy')
    search_dentists_in_city(city_name, 'social_facility')

print(len(df_other_healthcare))
df_other_healthcare.head(3)
```

```python
def average_range(range_str):
    # Remove dollar signs and commas, replace ASCII hyphen with special character, split, convert to integer and take average
    lower, upper = [int(x.replace('$', '').replace(',', '')) for x in range_str.replace('-', '−').split('−')]
    return (lower + upper) / 2
```

```python
def find_median_range(row, df):
    half_total = (row['Total'] - (row['Not applicable'] if 'Not applicable' in row else 0) - (row['Not stated'] if 'Not stated' in row else 0) - (row['Admin not available'] if 'Admin not available' in row else 0)) / 2
    cumulative_sum = 0
    for col in df.columns:
        if (col != 'LGA_NAME22' and col != 'Total' and col != 'Not applicable' and col != 'Not stated' and col != 'Admin not available'):
            cumulative_sum += row[col]
            if cumulative_sum >= half_total:
                return col
```

```python
df_lga_mortgage['score'] = 100 - 100 * (df_lga_mortgage['Median Value'] - df_lga_mortgage['Median Value'].min()) / (df_lga_mortgage['Median Value'].max() - df_lga_mortgage['Median Value'].min())
df_lga_mortgage = df_lga_mortgage.sort_values('score', ascending=False)
df_lga_mortgage = df_lga_mortgage.reset_index(drop=True)

print(len(df_lga_mortgage))
df_lga_mortgage.head(3)
```

```python
# Add score in Ranking table
df_indicators_NSW = addIndicatorScore('Mortgage', df_lga_mortgage, df_indicators_NSW)
df_indicators_NSW
```

```python
fig, ax = plt.subplots(figsize=(25,5))
plt.bar(df_lga_mortgage['LGA_NAME22'], df_lga_mortgage['score'])
plt.title('Mortgage Payment Score Ranking in NSW', fontsize = 15)
plt.xticks(rotation=90)
plt.show()
```

# 6: Policy Reflection

By applying EFI indicators, we can provide a rich information base and evidence base for policy. By using the EFI index, policies can be more targeted to solve the challenges brought by the future ageing of urban society. Measures can be formulated according to the ranking results and index analysis to change the direction of future urban planning and the areas where funds are tilted. At the same time, these indicators can also enable the government to better improve and optimise the existing urban social support services for the elderly, so as to provide better quality of life and well-being for the elderly. Make improvements in the following four aspects for the construction of equality policy in cities:

1. **Upholding Civic Responsibility:** Cities bear an intrinsic obligation to promote the well-being of all citizens, with the establishment of an elderly-friendly environment playing a paramount role in fostering social equality.

2. **Navigating Demographic Shifts:** Cities worldwide are grappling with a substantial demographic challenge - an ageing population. Those that can better adapt to the specific needs of older residents stand a greater chance of adeptly managing this trend.

3. **Enhancing Urban Appeal:** The allure of an elderly-friendly city transcends age brackets. Such cities don't merely appeal to older residents; they also draw in the younger generation, particularly those attentive to the well-being of their elderly kin.

4. **Driving Long-Term Urban Development:** An elderly-friendly city, typically characterised by robust public and medical infrastructures, is often indicative of a city on a sustainable developmental trajectory.

These points underline the gravity of this study, offering valuable insights to city planners and policy makers to ensure that urban spaces are adequately equipped to cater to an ageing population.

In the whole process of ranking, we found that due to the low population in remote areas, it is extremely easy to get a high score in terms of facilities per capita. However, the score of this part has a great influence on the ranking due to the high weight, so the weight allocation still needs to be improved in the future index adjustment.
At the same time, the above series of indicators are limited by the availability of data and the difficulties of statistics, and the consideration of factors such as human and natural environment and traffic still lacks some parts, which need to be further summarised and improved:

First, leisure activities policy, such as considering the range of activities for the elderly in the region and activity facilities. Whether the government has any consideration for the elderly. Second, in terms of transportation, whether there are barrier-free facilities for the elderly with mobility difficulties. Third, education and training to encourage the elderly to continue learning and development, such as free or preferential provision of educational resources for the elderly. 4. Natural environment, suitable for the elderly to live in the environment, green space, parks, etc., and ensure that the walking path and parks are easy to move around, considering the eyesight and mobility of the elderly. 5. Social security, such as traffic security, crime reduction, fraud prevention.
